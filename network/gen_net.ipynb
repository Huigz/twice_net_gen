```python
import networkx as nx
import numpy as np

def gini_coefficient(values):
    # 确保值是非负数
    values = np.array(values)
    if np.amin(values) < 0:
        values -= np.amin(values)
    
    # 排序
    values = np.sort(values)
    
    # 计算系数
    n = len(values)
    cumulative_sum = np.cumsum(values)
    gini = (n + 1 - 2 * np.sum(cumulative_sum) / cumulative_sum[-1]) / n
    
    return gini
```

```python
comp_net_degree = pd.DataFrame({
    'value': twice_all_degree_num + once_degree_num,
    'category': ['twice'] * len(twice_all_degree_num) + ['once'] * len(once_degree_num)
})
comp_net_degree
```

```python
def lorenz_curve(X_list):
    fig, ax = plt.subplots(figsize=[6,6])
    colors = ['blue', 'darkgreen']
    i=0
    for X in X_list:
        X = np.array(sorted(X))
        X_lorenz = X.cumsum() / X.sum()
        X_lorenz = np.insert(X_lorenz, 0, 0)
        X_lorenz[0], X_lorenz[-1]
        ax.scatter(np.arange(X_lorenz.size)/(X_lorenz.size-1), X_lorenz,
                marker='x', color=colors[i], s=1)
        i+=1
    ax.plot([0,1], [0,1], color='k')
```

```python
lorenz_curve([comp_net_degree[comp_net_degree['category']=='twice'].value.values, comp_net_degree[comp_net_degree['category']=='once'].value.values])
```
