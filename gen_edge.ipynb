```python
import pandas as pd
```

```python
once_post_df = pd.read_csv("once/once.csv")
once_comment_df = pd.read_csv("once/comment.csv")
once_recomment_df = pd.read_csv("once/recomment.csv")
```

```python
post_comment_df = pd.merge(left=once_post_df, right=once_comment_df, on="m_post_id", how="left")[['contents','user_id_x','user_id_y', 'm_comment_id']]
post_comment_df
```

```python
all_edge_df = pd.merge(left=post_comment_df, right=once_recomment_df, left_on="m_comment_id", right_on="from_comment_id",how="left")
all_edge_df = all_edge_df[['user_id_x', 'user_id_y', 'user_id']]
all_edge_df
```

```python
all_edge_df.to_csv("once_all_edge.csv", index=False)
```

# twice edge

```python
twice_post_df = pd.read_csv("twice/twice_post_clean.csv")
twice_comment_df = pd.read_csv("twice/comment.csv")
twice_recomment_df = pd.read_csv("twice/recomment.csv")
```

```python
twice_post_comment_df = pd.merge(left=twice_post_df, right=twice_comment_df, on="m_post_id", how="left")[['contents','user_id_x','user_id_y', 'm_comment_id']]
twice_post_comment_df
```

```python
twice_all_edge_df = pd.merge(left=twice_post_comment_df, right=twice_recomment_df, left_on="m_comment_id", right_on="from_comment_id",how="left")
twice_all_edge_df = twice_all_edge_df[['user_id_x', 'user_id_y', 'user_id']]
twice_all_edge_df
```

```python
twice_all_edge_df.to_csv("twice_all_edge.csv", index=False)
```
